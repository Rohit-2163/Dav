#### <span style="color:#007f5f">Create  a dataframe having at least 3 columns and 50 rows to store a numeric data generated by using random function. Replace 10% of the values by null values whose index position are generated using random function. Do the following </span>

```python
# importing numpy library
# importing pandas library

import numpy as np
import pandas as pd 
```

##### <span style="color:#003f88">a. Identify and count missing values in a dataframe</span>

```python
#creating a dataframe with 3 columns and 50 rows
df=pd.DataFrame(np.random.randint(0,100,size=(50,3)),
                columns=['col1','col2','col3'])

# Replacing 10% of the values with NaN
df=df.mask(np.random.random(df.shape)<.1)

#Identifying and count missing values in dataframe
missing_values=df.isnull().sum()

print('Missing Value in DataFrame is\n',missing_values)
print('Total Missing value ',missing_values.sum())
```

##### <span style="color:#003f88">b. Drop the column having more than 5 null values</span>


```python
df = df.dropna(thresh=len(df)-5, axis=1)
df
```

##### <span style="color:#003f88">c. Identify the row label having maximum of the sum of all values in a row and drop that row</span>

```python
# Identify the row label having maximum of the sum of all values in a row and drop that row
max_row_label = df.sum(axis=1).idxmax()
print("Dropped Row no : ",max_row_label,"having sum : ",df.sum(axis=1).max())
df = df.drop(max_row_label)
```

##### <span style="color:#003f88">d. Sort the dataframe on the basis of the first column.</span>

```python
# Sort the dataframe on the basis of the first column
df = df.sort_values(by=df.columns[0])
print("After sorting:")
df.head()
```

##### <span style="color:#003f88">e. Remove all duplicates from the first column.</span>

```python
# Remove all duplicates from the first column
df = df.drop_duplicates(subset=df.columns[0])
df.head()
```

##### <span style="color:#003f88">f.Find the correlation between first and second column and covariance between second and third column</span>

```python
correlation = df[df.columns[0]].corr(df[df.columns[1]])
covariance = df[df.columns[0]].cov(df[df.columns[1]])
print("Correlation : ",correlation)
print("Covariance  : ",covariance)
```

##### <span style="color:#003f88">g. Detect the outliers and remove the rows having outliers.</span>

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df=df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.head()
```

##### <span style="color:#003f88">h. Discretize second column and create 5 bins</span>

```python
# Discretize second column and create 5 bins
bbins=[0,20,40,60,80,100]
df[df.columns[1]] = pd.cut(df[df.columns[1]], bins=bbins)
df[df.columns[1]]
```
